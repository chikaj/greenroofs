```python
import rasterio as r
import geopandas as gpd
from rasterstats import zonal_stats
```

```python
footprints = gpd.read_file("data/greenroofs.gpkg", layer='Building Footprints')
```

```python
all_touched_false_2009 = zonal_stats(footprints, 'data/LT05_CU_027009_20090512_20210430_02_ST_B6.TIF')
all_touched_true_2009 = zonal_stats(footprints, 'data/LT05_CU_027009_20090512_20210430_02_ST_B6.TIF', all_touched=True)
all_touched_false_2019 = zonal_stats(footprints, 'data/LC08_CU_027009_20190524_20210504_02_ST_B10.TIF')
all_touched_true_2019 = zonal_stats(footprints, 'data/LC08_CU_027009_20190524_20210504_02_ST_B10.TIF', all_touched=True)
```

```python
meanT2009 = [i['mean'] for i in all_touched_true_2009]
meanT2019 = [i['mean'] for i in all_touched_true_2019]
```

```python
footprints
```

```python
footprints['meanT2009'] = meanT2009
footprints['meanT2019'] = meanT2019
```

```python
footprints.to_file("data/greenroofsWithTemperature.gpkg", layer='Building Footprints with Temperature', driver='GPKG')
```
